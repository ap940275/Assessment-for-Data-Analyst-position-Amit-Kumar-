```python
dict_data_list = list()
for gg, dd in df.groupby(['gender','specialisation']):
    group = dict(zip(['gender','specialisation'], gg))
    ocolumns_list = list()
    for _, data in dd.iterrows():
        data = data.drop(labels=['gender','specialisation'])
        ocolumns_list.append(data.to_dict())
    group['other_columns'] = ocolumns_list
    dict_data_list.append(group)

dict_data_list
```
